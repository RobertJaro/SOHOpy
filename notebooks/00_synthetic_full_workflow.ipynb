# SOHOpy LASCO Synthetic Full Workflow

This notebook is the friendly, no-network starting point. It creates synthetic LASCO-like inputs, runs the Level 1 reducer, writes tB/pB polarization products, and generates PNG previews.

Use the text box below to choose an output directory, then press **Run synthetic workflow**.

```python
from pathlib import Path

import ipywidgets as widgets
from IPython.display import Image, display

from examples.lasco_full_workflow import run_workflow
```

## Output Settings

```python
output_dir = widgets.Text(
    value="runs/notebook_synthetic_workflow",
    description="Output dir",
    layout=widgets.Layout(width="680px"),
)
run_button = widgets.Button(description="Run synthetic workflow", button_style="success")
run_output = widgets.Output()

display(widgets.VBox([output_dir, run_button, run_output]))
```

```python
def run_clicked(_):
    run_output.clear_output()
    with run_output:
        outputs = run_workflow(Path(output_dir.value).expanduser())
        print("Workflow complete. Products:")
        for name, path in outputs.items():
            print(f"  {name:16s} {path}")
        print("\nLevel 1 preview:")
        display(Image(filename=str(outputs["level1_png"])))
        print("\nPolarization preview:")
        display(Image(filename=str(outputs["polarization_png"])))

run_button.on_click(run_clicked)
```

## What To Check

The output directory should contain:

- a synthetic Level 0.5-like FITS input;
- a synthetic calibration cache;
- a Level 1 FITS product;
- a Level 1 PNG preview;
- a synthetic `-60/0/+60` polarizer triplet;
- tB and pB FITS products;
- a three-panel polarization summary PNG.
